In this notebook, we combine the median household income data files into one data file.

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#reading in data files 
A=pd.read_csv('Median_household_income_starts_with_A.csv',index_col='Category')
BC=pd.read_csv('Median_household_income_starts_with_B_OR_C.csv',index_col='Category')
DEF=pd.read_csv('Median_household_income_starts_with_D_E_F.csv',index_col='Category')
GH=pd.read_csv('Median_household_income_starts_with_G_H.csv',index_col='Category')
IJKL=pd.read_csv('Median_household_income_starts_with_I_J_K_L.csv',index_col='Category')
MN=pd.read_csv('Median_household_income_starts_with_M_N.csv',index_col='Category')
OPR=pd.read_csv('Median_household_income_starts_with_O_P_R.csv',index_col='Category')
SU=pd.read_csv('Median_household_income_starts_with_S_U.csv',index_col='Category')
W=pd.read_csv('Median_household_income_starts_with_W.csv',index_col='Category')
```

```python
#Concatenating dataframes
df_list=[A,BC,DEF,GH,IJKL,MN,OPR,SU,W]
median_household_income=pd.concat(df_list,axis=1)
median_household_income
```

```python
#Exporting data to a single csv
median_household_income.to_csv('median_household_income_merged.csv')
```
