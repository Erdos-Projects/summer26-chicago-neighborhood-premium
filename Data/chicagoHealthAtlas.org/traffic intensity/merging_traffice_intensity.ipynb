In this notebook, we combine the traffic intensity data files into one data file.

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#reading in data files 
A=pd.read_csv('traffic_intensity_starts_with_A.csv',index_col='Category')
BC=pd.read_csv('traffic_intensity_starts_with_B_C.csv',index_col='Category')
DEFG=pd.read_csv('traffic_intensity_starts_with_D_E_F_G.csv',index_col='Category')
HIJK=pd.read_csv('traffic_intensity_starts_with_H_I_J_K.csv',index_col='Category')
LM=pd.read_csv('traffic_intensity_starts_with_L_M.csv',index_col='Category')
NO=pd.read_csv('traffic_intensity_starts_with_N_O.csv',index_col='Category')
PRSU=pd.read_csv('traffic_intensity_starts_with_P_R_S_U.csv',index_col='Category')
W=pd.read_csv('traffic_intensity_starts_with_W.csv',index_col='Category')
```

```python
#Concatenating dataframes
df_list=[A,BC,DEFG,HIJK,LM,NO,PRSU,W]
traffic_intensity=pd.concat(df_list,axis=1)
traffic_intensity
```

```python
#Exporting data to a single csv
traffic_intensity.to_csv('traffic_intensity_merged.csv')
```
