In this notebook, we combine the employee-based health insurance data files into one data file. 

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#reading in data files 
A=pd.read_csv('Employee_based_health_insurance_starts_with_A.csv',index_col='Category')
BC=pd.read_csv('Employee_based_health_insurance_starts_with_B_C.csv',index_col='Category')
DEF=pd.read_csv('Employee_based_health_insurance_starts_with_D_E_F.csv',index_col='Category')
GHI=pd.read_csv('Employee_based_health_insurance_starts_with_G_H_I.csv',index_col='Category')
JKL=pd.read_csv('Employee_based_health_insurance_starts_with_J_K_L.csv',index_col='Category')
MN=pd.read_csv('Employee_based_health_insurance_starts_with_M_N.csv',index_col='Category')
OPR=pd.read_csv('Employee_based_health_insurance_starts_with_O_P_R.csv',index_col='Category')
SU=pd.read_csv('Employee_based_health_insurance_starts_with_S_U.csv',index_col='Category')
W=pd.read_csv('Employee_based_health_insurance_starts_with_W.csv',index_col='Category')
```

```python
#Concatenating dataframes
df_list=[A,BC,DEF,GHI,JKL,MN,OPR,SU,W]
employee_based_health_ins=pd.concat(df_list,axis=1)
employee_based_health_ins
```

```python
employee_based_health_ins.info()
```

```python
#Exporting data to a single csv
employee_based_health_ins.to_csv('employee_based_health_insurance_merged.csv')
```
