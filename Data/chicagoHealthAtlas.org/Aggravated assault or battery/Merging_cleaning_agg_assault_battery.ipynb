In this notebook, we combine the aggravated assault and battery data files into one data file.

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#reading in data files 
A=pd.read_csv('Aggravated_assault_battery_starts_with_A.csv',index_col='Category')
BC=pd.read_csv('Aggravated_assault_battery_starts_with_B_C.csv',index_col='Category')
DEF=pd.read_csv('Aggravated_assault_battery_starts_with_D_E_F.csv',index_col='Category')
GHIJ=pd.read_csv('Aggravated_assault_battery_starts_with_G_H_I_J.csv',index_col='Category')
KLM=pd.read_csv('Aggravated_assault_battery_starts_with_K_L_M.csv',index_col='Category')
NO=pd.read_csv('Aggravated_assault_battery_starts_with_N_O.csv',index_col='Category')
PRSU=pd.read_csv('Aggravated_assault_battery_starts_with_P_R_S_U.csv',index_col='Category')
W=pd.read_csv('Aggravated_assault_battery_starts_with_W.csv',index_col='Category')
```

```python
#Concatenating dataframes
df_list=[A,BC,DEF,GHIJ,KLM,NO,PRSU,W]
agg_assault_battery=pd.concat(df_list,axis=1)
agg_assault_battery
```

```python
agg_assault_battery.info()
```

```python
#Exporting data to a single csv
agg_assault_battery.to_csv('agg_assault_battery_merged.csv')
```
