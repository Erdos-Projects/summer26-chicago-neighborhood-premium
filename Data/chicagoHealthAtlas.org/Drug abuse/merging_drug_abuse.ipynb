In this notebook, we combine the drug abuse data files into one data file. 

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#reading in data files 
A=pd.read_csv('drug_abuse_starts_with_A.csv',index_col='Category')
BCD=pd.read_csv('drug_abuse_starts_with_B_C_D.csv',index_col='Category')
EFG=pd.read_csv('drug_abuse_starts_with_E_F_G.csv',index_col='Category')
HIJK=pd.read_csv('drug_abuse_starts_with_H_I_J_K.csv',index_col='Category')
LM=pd.read_csv('drug_abuse_starts_with_L_M.csv',index_col='Category')
NO=pd.read_csv('drug_abuse_starts_with_N_O.csv',index_col='Category')
PRSU=pd.read_csv('drug_abuse_starts_with_P_R_S_U.csv',index_col='Category')
W=pd.read_csv('drug_abuse_starts_with_W.csv',index_col='Category')
```

```python
#Concatenating dataframes
df_list=[A,BCD,EFG,HIJK,LM,NO,PRSU,W]
drug_abuse=pd.concat(df_list,axis=1)
drug_abuse
```

```python
drug_abuse.info()
```

```python
#Exporting data to a single csv
drug_abuse.to_csv('drug_abuse_merged.csv')
```
