```python
#import packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import Zillow neighborhood data.

Zillow=pd.read_csv('Neighborhood_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv')
Zillow.head(5)
```

The data above was found at 'https://www.zillow.com/research/data/'. It comes from Home values using the Zillow Home Value Index (ZHVI) calculated using the neural Zestimate. 
Neural Zestimate information found here: 'https://www.zillow.com/research/methodology-neural-zhvi-32128/'


```python
#Retrieve data about Chicago, IL
Zillow_Chicago=Zillow[(Zillow.State=='IL')&(Zillow.City=='Chicago')]
Zillow_Chicago
```

```python
#Exporting the dataframe above as a csv file
Zillow_Chicago.to_csv('Zillow_Chicago_ZHVI_by_neighborhood.csv',index=False)
```

```python
Zillow_Chicago.groupby(by='RegionName').value_counts()
```

Each row in the dataframe above is a time series for the ZHVI each month for one neighborhood in Chicago. Each month's number is the typical home value, 'typical' meaning between the 35th and 65th percentile.
