In this notebook, we combine the traffic intensity data files into one data file.

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#reading in data files 
A=pd.read_csv('unemployment_rate_starts_with_A.csv',index_col='Category')
BC=pd.read_csv('unemployment_rate_starts_with_B_C.csv',index_col='Category')
DEF=pd.read_csv('unemployment_rate_starts_with_D_E_F.csv',index_col='Category')
GHIJ=pd.read_csv('unemployment_rate_starts_with_G_H_I_J.csv',index_col='Category')
KL=pd.read_csv('unemployment_rate_starts_with_K_L.csv',index_col='Category')
MN=pd.read_csv('unemployment_rate_starts_with_M_N.csv',index_col='Category')
OPR=pd.read_csv('unemployment_rate_starts_with_O_P_R.csv',index_col='Category')
SU=pd.read_csv('unemployment_rate_starts_with_S_U.csv',index_col='Category')
W=pd.read_csv('unemployment_rate_starts_with_W.csv',index_col='Category')
```

```python
#Concatenating dataframes
df_list=[A,BC,DEF,GHIJ,KL,MN,OPR,SU,W]
unemployment_rate=pd.concat(df_list,axis=1)
unemployment_rate
```

```python
#Exporting data to a single csv
unemployment_rate.to_csv('unemployment_rate_merged.csv')
```
