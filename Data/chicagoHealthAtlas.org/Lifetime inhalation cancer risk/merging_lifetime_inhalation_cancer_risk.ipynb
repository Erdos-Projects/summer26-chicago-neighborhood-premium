In this notebook, we combine the lifetime inhalation cancer risk data files into one data file.

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#reading in data files 
A=pd.read_csv('lifetime_inhalation_cancer_risk_starts_with_A.csv',index_col='Category')
BC=pd.read_csv('lifetime_inhalation_cancer_risk_starts_with_B_C.csv',index_col='Category')
DEF=pd.read_csv('lifetime_inhalation_cancer_risk_starts_with_D_E_F.csv',index_col='Category')
GHIJK=pd.read_csv('lifetime_inhalation_cancer_risk_starts_with_G_H_I_J_K.csv',index_col='Category')
LM=pd.read_csv('lifetime_inhalation_cancer_risk_starts_with_L_M.csv',index_col='Category')
NO=pd.read_csv('lifetime_inhalation_cancer_risk_starts_with_N_O.csv',index_col='Category')
PRSU=pd.read_csv('lifetime_inhalation_cancer_risk_starts_with_P_R_S_U.csv',index_col='Category')
W=pd.read_csv('lifetime_inhalation_cancer_risk_starts_with_W.csv',index_col='Category')
```

```python
#Concatenating dataframes
df_list=[A,BC,DEF,GHIJK,LM,NO,PRSU,W]
lifetime_inhalation_cancer_risk=pd.concat(df_list,axis=1)
lifetime_inhalation_cancer_risk
```

We see that every neighborhood is missing data for the same years. We need to decide whether to impute data, drop these years, or abandon this data set entirely.

```python
#Exporting data to a single csv
lifetime_inhalation_cancer_risk.to_csv('lifetime_inhalation_cancer_risk_merged.csv')
```
