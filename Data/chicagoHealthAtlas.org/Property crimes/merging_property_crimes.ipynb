In this notebook, we combine the property crime data files into one data file.

```python
#importing packages
import pandas as pd
import numpy as np
```

```python
#reading in data files 
A=pd.read_csv('Property_crimes_starts_with_A.csv',index_col='Category')
BC=pd.read_csv('Property_crimes_starts_with_B_C.csv',index_col='Category')
DEF=pd.read_csv('Property_crimes_starts_with_D_E_F.csv',index_col='Category')
GHIJ=pd.read_csv('Property_crimes_starts_with_G_H_I_J.csv',index_col='Category')
KLM=pd.read_csv('Property_crimes_starts_with_K_L_M.csv',index_col='Category')
NO=pd.read_csv('Property_crimes_starts_with_N_O.csv',index_col='Category')
PRSU=pd.read_csv('Property_crimes_starts_with_P_R_S_U.csv',index_col='Category')
W=pd.read_csv('Property_crimes_starts_with_W.csv',index_col='Category')
```

```python
#Concatenating dataframes
df_list=[A,BC,DEF,GHIJ,KLM,NO,PRSU,W]
property_crimes=pd.concat(df_list,axis=1)
property_crimes
```

```python
#Exporting data to a single csv
property_crimes.to_csv('property_crimes_merged.csv')
```
